```python
def mRNAvProteinplot_scatter(df, x_col='MeanExpmRNA', y_col='MeanExpmRNA', color_col='Condition'):
    fig, ax = plt.subplots(figsize=(8, 6))

    for condition in combined_df[color_col].unique():
        subset = df[df[color_col] == condition]
        ax.scatter(subset[x_col], subset[y_col], label=condition, alpha=0.7)

    ax.set_xlabel(x_col)
    ax.set_ylabel(y_col)
    ax.set_title(f'{y_col} vs {x_col}')

    ax.legend(title=color_col)
    ax.grid(True)
    fig.tight_layout()
    plt.show()
```
